```python
import numpy as np
import csv
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
name1='./Episode_400_K_1.summary_csv'
name2='./Episode_400_K_2.summary_csv'
name3='./Episode_400_K_3.summary_csv'
name5='./Episode_400_K_5.summary_csv'
name10='./Episode_400_K_10.summary_csv'
data1=np.loadtxt(name1,delimiter=',')
data2=np.loadtxt(name2,delimiter=',')
data3=np.loadtxt(name3,delimiter=',')
data5=np.loadtxt(name5,delimiter=',')
data10=np.loadtxt(name10,delimiter=',')
```

# The error for estimating Q-function

```python
plt.plot(data1[:,1],color='red',label='K=1')
plt.plot(data2[:,1],color='blue',label='K=2')
plt.plot(data3[:,1],color='green',label='K=3')
plt.plot(data5[:,1],color='black',label='K=5')
plt.plot(data10[:,1],color='cyan',label='K=10')
plt.legend()
plt.title("Estimation error for Q-function")
plt.ylim(0,100)
plt.xlabel('Episode')
plt.ylabel("Error")
plt.show()
```

# Average Error for K

```python
print('K=1',np.mean(data1[np.where(data1[:,1]!=-1)[0],1]))
print('K=2',np.mean(data2[np.where(data2[:,1]!=-1)[0],1]))
print('K=3',np.mean(data3[np.where(data3[:,1]!=-1)[0],1]))
print('K=５',np.mean(data5[np.where(data5[:,1]!=-1)[0],1]))
print('K=10',np.mean(data10[np.where(data10[:,1]!=-1)[0],1]))
```

# Reward against the episode

```python
plt.plot(data1[:,2],color='red',label='K=1')
plt.plot(data2[:,2],color='blue',label='K=2')
plt.plot(data3[:,2],color='green',label='K=3')
plt.plot(data5[:,2],color='black',label='K=5')
plt.plot(data10[:,2],color='cyan',label='K=10')

plt.legend()
plt.title("Reward")
plt.ylim(0,210)
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.show()
```

```python
print('K=1',np.mean(data1[np.where(data1[:,1]!=-1)[0],2]))
print('K=2',np.mean(data2[np.where(data2[:,1]!=-1)[0],2]))
print('K=3',np.mean(data3[np.where(data3[:,1]!=-1)[0],2]))
print('K=5',np.mean(data5[np.where(data5[:,1]!=-1)[0],2]))
print('K=10',np.mean(data10[np.where(data10[:,1]!=-1)[0],2]))
```
